## Building a Logistic Regression with Cost Function and Gradient Descent

```python
theta = np.zeros((X.shape[1], 1))
```

```python
def sigmoid(x):
    # Activation function used to map any real value between 0 and 1
    return 1 / (1 + np.exp(-x))

def net_input(theta, x):
    # Computes the weighted sum of inputs
    return np.dot(x, theta)

def probability(theta, x):
    # Returns the probability after passing through sigmoid
    return sigmoid(net_input(theta, x))
```

```python
def cost_function(theta, x, y):
    # Computes the cost function for all the training samples
    m = x.shape[0]
    total_cost = -(1 / m) * np.sum(y * np.log(probability(theta, x)) + (1 - y) * np.log(1 - probability(theta, x)))
    return total_cost

def gradient(theta, x, y):
    # Computes the gradient of the cost function at the point theta
    m = x.shape[0]
    return (1 / m) * np.dot(x.T, sigmoid(net_input(theta,x)) - y)
```

```python
# from scipy.optimize import fmin_tnc
# def fit(x, y, theta):
#     opt_weights = fmin_tnc(func=cost_function, x0=theta, fprime=gradient, args=(x, y))
#     return opt_weights[0]
# parameters = fit(X, y, theta)


import numpy as np
from scipy.optimize import fmin_tnc

# 1. Sigmoid function
def sigmoid(z):
    z = np.array(z)  # ensures z supports vectorized np.exp
    return 1 / (1 + np.exp(-z))

# 2. Cost function
def cost_function(theta, X, y):
    h = sigmoid(np.dot(X, theta))
    epsilon = 1e-5  # avoid log(0)
    return -np.mean(y * np.log(h + epsilon) + (1 - y) * np.log(1 - h + epsilon))

# 3. Gradient of the cost function
def gradient(theta, X, y):
    h = sigmoid(np.dot(X, theta))
    return np.dot(X.T, (h - y)) / y.size

# 4. Fit function using scipy fmin_tnc
def fit(X, y, theta):
    theta = np.array(theta, dtype=np.float64)
    return fmin_tnc(func=cost_function, x0=theta, fprime=gradient, args=(X, y))[0]

# 5. Example usage:
# Ensure X is a NumPy array
X = np.array(X, dtype=np.float64)

# Add intercept (bias) term
X = np.c_[np.ones((X.shape[0], 1)), X]

# Ensure y is a 1D NumPy array of floats
y = np.array(y, dtype=np.float64).flatten()

# Initialize theta to zeros
theta = np.zeros(X.shape[1])

# Train model
theta_optimized = fit(X, y, theta)
```

```python
parameters = theta_optimized
```

```python
def predict(x):
    theta = parameters[:, np.newaxis]
    return probability(theta, x)
def accuracy(x, actual_classes, probab_threshold=0.5):
    predicted_classes = (predict(x) >= 
                         probab_threshold).astype(int)
    predicted_classes = predicted_classes.flatten()
    accuracy = np.mean(predicted_classes == actual_classes)
    return accuracy * 100
accuracy(X, y)
```
